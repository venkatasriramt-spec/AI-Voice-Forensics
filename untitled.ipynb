# Spectrogram Data Generation

```python
"""
=============================================================================
PIPELINE STEP 1: PARALLEL AUDIO PREPROCESSOR
=============================================================================
PURPOSE:
To convert raw `.wav` audio files into deep-learning-ready Mel-Spectrogram 
matrices (`.npy` files) via a high-speed, parallel CPU execution pool.

WHY THIS IS NECESSARY:
Neural Networks (specifically CNNs) cannot process raw `.wav` files efficiently 
during training. If the model had to convert audio into spectrograms "on the fly" 
during every single epoch, the training loop would be bottlenecked by CPU overhead. 
This script performs that heavy lifting *once*, offline, dumping mathematically 
perfect arrays to the disk.
=============================================================================
"""
import os
import librosa
import numpy as np
import glob
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm import tqdm

DATA_DIR = './local_data/'
OUTPUT_DIR = './spectrogram_data/'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 1. The conversion function
def process_audio(file_path):
    try:
        y, sr = librosa.load(file_path, sr=16000, duration=2.5)
        mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
        mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
        
        # Keep the exact same Language/Real/Fake folder structure
        rel_path = os.path.relpath(file_path, DATA_DIR)
        out_path = os.path.join(OUTPUT_DIR, rel_path).replace('.wav', '.npy')
        
        os.makedirs(os.path.dirname(out_path), exist_ok=True)
        np.save(out_path, mel_spec_db)
        
        # Return success and no error
        return (True, None)
    except Exception as e:
        # Return failure and the exact error message
        return (False, f"Failed on {file_path}: {str(e)}")

# 2. The main execution block
if __name__ == '__main__':
    all_wav_files = glob.glob(os.path.join(DATA_DIR, '**/*.wav'), recursive=True)
    print(f"Found {len(all_wav_files)} audio files. Spinning up CPU cores...")
    
    successful_conversions = 0
    errors = []
    
    # 3. Process all files across all cores with a clean progress bar
    with ProcessPoolExecutor() as executor:
        # Submit all tasks to the cores
        futures = {executor.submit(process_audio, path): path for path in all_wav_files}
        
        # As each file finishes, update the progress bar
        for future in tqdm(as_completed(futures), total=len(all_wav_files), desc="Generating Spectrograms", unit="file"):
            success, error_msg = future.result()
            if success:
                successful_conversions += 1
            else:
                errors.append(error_msg)
                
    # 4. Print the final summary and any collected errors at the very bottom
    print("\n" + "="*40)
    print("CONVERSION JOB FINISHED")
    print("="*40)
    print(f"Successfully converted: {successful_conversions} / {len(all_wav_files)}")
    
    if errors:
        print(f"\nEncountered {len(errors)} errors during processing:")
        for err in errors:
            print(f" - {err}")
    else:
        print("\nFlawless execution. Zero errors encountered.")
```

# Spectrogram Data Preparation for Training

```python
"""
=============================================================================
PIPELINE STEP 2: PYTORCH DATASET INGESTION MODULE
=============================================================================
PURPOSE:
Defines the `AudioForensicsDataset` class, which dictates exactly how the 
PyTorch DataLoader should stream the pre-processed `.npy` spectrograms 
into the GPU during training.

KEY OPERATIONS:
1. File Streaming: Loads the NumPy arrays dynamically from the disk.
2. Channel Expansion: Neural Networks expect images in (Channels, Height, Width) 
   format. Because our spectrograms are 2D, this script manually injects a 
   dimension (`np.expand_dims`) so PyTorch treats it as a 1-Channel (Grayscale) image.
3. Auto-Labeling: Parses the directory string; if the array originated from a 
   "Fake Audio" folder, it attaches a Ground Truth label of 1.0 (Synthetic).
=============================================================================
"""
import os
import glob
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader

class AudioForensicsDataset(Dataset):
    def __init__(self, data_dir):
        # Find all the new .npy spectrogram files
        self.files = glob.glob(os.path.join(data_dir, '**/*.npy'), recursive=True)
        print(f"Dataset loaded! Found {len(self.files)} spectrograms ready for training.")
        
    def __len__(self):
        return len(self.files)
        
    def __getitem__(self, idx):
        file_path = self.files[idx]
        
        # 1. Load the spectrogram array
        spec = np.load(file_path)
        
        # 2. PyTorch expects images to have a "color channel". 
        # Since our spectrogram is basically grayscale, we add a 1-channel dimension.
        spec = np.expand_dims(spec, axis=0) 
        spec_tensor = torch.tensor(spec, dtype=torch.float32)
        
        # 3. Label it! If it's in the "Fake Audio" folder, it gets a 1. Otherwise, 0.
        if 'Fake Audio' in file_path:
            label = 1.0 # AI Clone
        else:
            label = 0.0 # Real Human
            
        return spec_tensor, torch.tensor([label], dtype=torch.float32)

# Create the dataset object pointing to your new folder
dataset = AudioForensicsDataset('./spectrogram_data/')
```

# CRNN Original Brain (Version 1)

```python
"""
=============================================================================
THUNDERBUDDIES CORE ARCHITECTURE & HARDWARE VALIDATION
=============================================================================
WHAT THIS SCRIPT DOES:
This script defines the foundational neural network blueprint for the entire 
Audio Forensics project: a hybrid CRNN (Convolutional Recurrent Neural Network).
It stitches together a modified ResNet18 (to "see" 1-channel spectrogram 
images) and a Bidirectional LSTM (to "remember" acoustic sequences over time). 

Crucially, it includes a hardware-level override (`cudnn.enabled = False`) 
to bypass a known NVIDIA library crashing bug, allowing the model to safely 
mount onto your GPU.

IMPORTANCE FOR V2 & V3 TRAINING SCRIPTS:
1. The Core Blueprint: Both the V2 (Aggressive Augmentation) and V3 (Hybrid 
   Augmentation) training scripts are essentially just "classrooms." THIS 
   script defines the actual "student" (`ThunderBuddiesCRNN`). Both V2 and 
   V3 completely rely on this exact class definition to exist.
   
2. The Diagnostic Sanity Check: Before kicking off a massive, multi-hour 
   V2 or V3 training loop, running this script acts as a safe diagnostic test. 
   If this script successfully prints "Brain successfully built and moved to: cuda", 
   it proves your PyTorch environment is stable, the GPU is communicating, 
   and V2/V3 are safe to execute without instantly crashing.
=============================================================================
"""
import torch
import torch.nn as nn
import torchvision.models as models

# --- THE MAGIC BYPASS ---
# This tells PyTorch to use the V100 GPU, but ignore the conflicting cuDNN library
torch.backends.cudnn.enabled = False 
# ------------------------

class ThunderBuddiesCRNN(nn.Module):
    def __init__(self):
        super(ThunderBuddiesCRNN, self).__init__()
        
        # 1. THE EYES (CNN)
        self.resnet = models.resnet18(pretrained=False)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.features = nn.Sequential(*list(self.resnet.children())[:-1])
        
        # 2. THE MEMORY (RNN)
        self.lstm = nn.LSTM(input_size=512, hidden_size=128, num_layers=1, batch_first=True, bidirectional=True)
        
        # 3. THE JUDGE (Classifier)
        self.fc = nn.Linear(256, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.features(x)                  
        x = x.squeeze(-1).squeeze(-1)         
        x = x.unsqueeze(1)                    
        lstm_out, _ = self.lstm(x)            
        x = lstm_out[:, -1, :]                
        x = self.fc(x)
        x = self.sigmoid(x)                   
        return x

# Move the brain to your powerful V100 GPU!
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ThunderBuddiesCRNN().to(device)

print("Brain successfully built and moved to:", device)
```

# Model Version 2

```python
"""
=============================================================================
AUDIO FORENSICS V2 - AGGRESSIVE AUGMENTATION TRAINING PIPELINE
=============================================================================
This script trains the 'ThunderBuddiesCRNN' deepfake detection model using 
a ResNet18 spatial feature extractor coupled with a bidirectional LSTM.

V2 STRATEGY (100% DEGRADATION):
This version utilizes an aggressive, deterministic data augmentation strategy 
running directly on the GPU. It injects static noise and applies SpecAugment 
(Time and Frequency masking) to EVERY SINGLE BATCH during training. 

WHY? By completely destroying the pristine acoustic fingerprints of the 
training data, the network cannot take the "lazy route" of just detecting 
how clean the audio is. It is forced to learn the mathematically rigid, 
underlying vocoder artifacts left behind by AI generation.
=============================================================================
"""
import os
import glob
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader, random_split
import torch.nn.functional as F
import torchaudio.transforms as T
from tqdm import tqdm

torch.backends.cudnn.enabled = False

# --- 1. THE DATASET LOADER (Pristine, no changes needed here) ---
class AudioForensicsDataset(Dataset):
    def __init__(self, data_dir):
        self.files = glob.glob(os.path.join(data_dir, '**/*.npy'), recursive=True)
        print(f"Loaded {len(self.files)} spectrogram arrays.")
        
    def __len__(self):
        return len(self.files)
        
    def __getitem__(self, idx):
        file_path = self.files[idx]
        spec = np.load(file_path)
        spec = np.expand_dims(spec, axis=0) 
        spec_tensor = torch.tensor(spec, dtype=torch.float32)
        
        TARGET_WIDTH = 80
        current_width = spec_tensor.shape[2]
        
        if current_width < TARGET_WIDTH:
            padding = TARGET_WIDTH - current_width
            spec_tensor = F.pad(spec_tensor, (0, padding))
        elif current_width > TARGET_WIDTH:
            spec_tensor = spec_tensor[:, :, :TARGET_WIDTH]
        
        label = 1.0 if 'Fake Audio' in file_path else 0.0
        return spec_tensor, torch.tensor([label], dtype=torch.float32)

# --- 2. THE MODEL ARCHITECTURE ---
class ThunderBuddiesCRNN(nn.Module):
    def __init__(self):
        super(ThunderBuddiesCRNN, self).__init__()
        self.resnet = models.resnet18(weights=None)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.features = nn.Sequential(*list(self.resnet.children())[:-1])
        self.lstm = nn.LSTM(input_size=512, hidden_size=128, num_layers=1, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(256, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.features(x)                  
        x = x.squeeze(-1).squeeze(-1)         
        x = x.unsqueeze(1)                    
        lstm_out, _ = self.lstm(x)            
        x = lstm_out[:, -1, :]                
        x = self.fc(x)
        x = self.sigmoid(x)                   
        return x

if __name__ == '__main__':
    dataset = AudioForensicsDataset('./spectrogram_data/')
    total_size = len(dataset)
    train_size = int(0.8 * total_size)
    val_size = total_size - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=8)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=8)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = ThunderBuddiesCRNN().to(device)
    criterion = nn.BCELoss() 
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # --- NEW: THE AUGMENTATION TOOLS (Loaded to GPU) ---
    freq_mask = T.FrequencyMasking(freq_mask_param=15).to(device)
    time_mask = T.TimeMasking(time_mask_param=10).to(device)
    # ---------------------------------------------------

    epochs = 5
    # --- NEW: Save as Version 2 ---
    save_path = "thunderbuddies_model_v2.pth"
    best_val_loss = float('inf') 

    print("\n" + "="*50)
    print(f"STARTING V2 ROBUST TRAINING ON V100 GPU")
    print(f"Training: {train_size} | Validation: {val_size}")
    print("="*50)

    for epoch in range(epochs):
        model.train()
        running_train_loss = 0.0
        correct_train = 0
        total_train = 0
        
        train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [Train]", leave=False)
        for inputs, labels in train_bar:
            inputs, labels = inputs.to(device), labels.to(device)
            
            # --- NEW: BULLETPROOF AUGMENTATION (Applied instantly on GPU) ---
            # 1. Inject static noise
            noise = torch.randn_like(inputs) * 0.05
            inputs = inputs + noise
            # 2. Apply Masks (Destroys vocoder fingerprints & simulates compression)
            inputs = freq_mask(inputs)
            inputs = time_mask(inputs)
            # ----------------------------------------------------------------
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_train_loss += loss.item()
            predictions = (outputs > 0.5).float()
            correct_train += (predictions == labels).sum().item()
            total_train += labels.size(0)
            train_bar.set_postfix(loss=(running_train_loss/len(train_loader)))

        model.eval()
        running_val_loss = 0.0
        correct_val = 0
        total_val = 0
        
        val_bar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{epochs} [Val  ]", leave=False)
        with torch.no_grad():
            for inputs, labels in val_bar:
                inputs, labels = inputs.to(device), labels.to(device)
                
                # NOTE: We specifically DO NOT augment the validation data here!
                # We want to test on the pristine, unedited clips.
                
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                running_val_loss += loss.item()
                predictions = (outputs > 0.5).float()
                correct_val += (predictions == labels).sum().item()
                total_val += labels.size(0)
                
        train_acc = correct_train / total_train
        val_acc = correct_val / total_val
        current_val_loss = running_val_loss / len(val_loader)
        
        print(f"Epoch {epoch+1} Summary | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | Val Loss: {current_val_loss:.4f}")

        if current_val_loss < best_val_loss:
            best_val_loss = current_val_loss
            torch.save(model.state_dict(), save_path)
            print(f"   🌟 New High Score! Model saved with Val Loss: {best_val_loss:.4f}")
        else:
            print(f"   ❌ Overfitting detected (Loss went up). Keeping previous best model.")

    print("\n" + "="*50)
    print("TRAINING COMPLETE. VERSION 2 SAVED.")
    print("="*50)
```

# Training Model Version 3

```python
"""
=============================================================================
AUDIO FORENSICS V3 - HYBRID / PROBABILISTIC TRAINING PIPELINE
=============================================================================
This script trains the 'ThunderBuddiesCRNN' deepfake detection model using 
a ResNet18 spatial feature extractor coupled with a bidirectional LSTM.

V3 STRATEGY (PROBABILISTIC + ANALOG HOLE SMEARING):
This is the most advanced, highly generalizable iteration of the training loop. 
It differs from V2 in two major, strategic ways:

1. Probabilistic Application (50%): 
   Augmentations are NOT applied to every batch. They trigger randomly 50% 
   of the time. This forces the model to learn from BOTH pristine, high-fidelity 
   studio AI audio AND heavily degraded VoIP audio, maximizing generalizability.
   
2. "Analog Hole" Simulation (Simulated Reverb): 
   V3 introduces a novel augmentation that shifts and decays the spectrogram 
   matrices. This mathematically simulates how synthetic sound waves echo 
   and bounce in a physical room before hitting a live microphone.
=============================================================================
"""
import os
import glob
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
from torch.utils.data import Dataset, DataLoader, random_split
import torch.nn.functional as F
import torchaudio.transforms as T
from tqdm import tqdm

torch.backends.cudnn.enabled = False

# --- 1. THE DATASET LOADER ---
class AudioForensicsDataset(Dataset):
    def __init__(self, data_dir):
        self.files = glob.glob(os.path.join(data_dir, '**/*.npy'), recursive=True)
        print(f"Loaded {len(self.files)} spectrogram arrays.")
        
    def __len__(self):
        return len(self.files)
        
    def __getitem__(self, idx):
        file_path = self.files[idx]
        spec = np.load(file_path)
        spec = np.expand_dims(spec, axis=0) 
        spec_tensor = torch.tensor(spec, dtype=torch.float32)
        
        TARGET_WIDTH = 80
        current_width = spec_tensor.shape[2]
        
        if current_width < TARGET_WIDTH:
            padding = TARGET_WIDTH - current_width
            spec_tensor = F.pad(spec_tensor, (0, padding))
        elif current_width > TARGET_WIDTH:
            spec_tensor = spec_tensor[:, :, :TARGET_WIDTH]
        
        label = 1.0 if 'Fake Audio' in file_path else 0.0
        return spec_tensor, torch.tensor([label], dtype=torch.float32)

# --- 2. THE MODEL ARCHITECTURE ---
class ThunderBuddiesCRNN(nn.Module):
    def __init__(self):
        super(ThunderBuddiesCRNN, self).__init__()
        self.resnet = models.resnet18(weights=None)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.features = nn.Sequential(*list(self.resnet.children())[:-1])
        self.lstm = nn.LSTM(input_size=512, hidden_size=128, num_layers=1, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(256, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.features(x)                  
        x = x.squeeze(-1).squeeze(-1)         
        x = x.unsqueeze(1)                    
        lstm_out, _ = self.lstm(x)            
        x = lstm_out[:, -1, :]                
        x = self.fc(x)
        x = self.sigmoid(x)                   
        return x

if __name__ == '__main__':
    dataset = AudioForensicsDataset('./spectrogram_data/')
    total_size = len(dataset)
    train_size = int(0.8 * total_size)
    val_size = total_size - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=8)
    val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=8)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = ThunderBuddiesCRNN().to(device)
    criterion = nn.BCELoss() 
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # The Visual Augmentation Tools
    freq_mask = T.FrequencyMasking(freq_mask_param=15).to(device)
    time_mask = T.TimeMasking(time_mask_param=10).to(device)

    epochs = 5
    # --- SAFEGUARD: Saving as a completely new Version 3 file ---
    save_path = "thunderbuddies_model_v3.pth"
    best_val_loss = float('inf') 

    print("\n" + "="*50)
    print(f"STARTING V3 (REVERB + NOISE) TRAINING ON V100 GPU")
    print(f"Training: {train_size} | Validation: {val_size}")
    print("="*50)

    for epoch in range(epochs):
        model.train()
        running_train_loss = 0.0
        correct_train = 0
        total_train = 0
        
        train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [Train]", leave=False)
        for inputs, labels in train_bar:
            inputs, labels = inputs.to(device), labels.to(device)
            
            # --- THE V3 COMBINED AUGMENTATION PIPELINE ---
            # Apply augmentations randomly (50% of the time) so it learns both clean and noisy environments
            if torch.rand(1).item() > 0.5:
                # 1. Analog Hole Smearing (Simulated Reverb)
                # Shifts the image right and stacks it to mimic echoing sound waves
                smear = inputs.clone()
                decay = 0.6
                for i in range(1, 5): 
                    shifted = F.pad(inputs[:, :, :, :-i], (i, 0))
                    smear += shifted * (decay ** i)
                inputs = smear
                
                # 2. Inject static noise (Mic fuzz)
                noise = torch.randn_like(inputs) * 0.05
                inputs = inputs + noise
                
                # 3. Apply Masks (Compression simulation)
                inputs = freq_mask(inputs)
                inputs = time_mask(inputs)
            # ---------------------------------------------
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_train_loss += loss.item()
            predictions = (outputs > 0.5).float()
            correct_train += (predictions == labels).sum().item()
            total_train += labels.size(0)
            train_bar.set_postfix(loss=(running_train_loss/len(train_loader)))

        model.eval()
        running_val_loss = 0.0
        correct_val = 0
        total_val = 0
        
        val_bar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{epochs} [Val  ]", leave=False)
        with torch.no_grad():
            for inputs, labels in val_bar:
                inputs, labels = inputs.to(device), labels.to(device)
                
                # Validation data remains totally pristine
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                running_val_loss += loss.item()
                predictions = (outputs > 0.5).float()
                correct_val += (predictions == labels).sum().item()
                total_val += labels.size(0)
                
        train_acc = correct_train / total_train
        val_acc = correct_val / total_val
        current_val_loss = running_val_loss / len(val_loader)
        
        print(f"Epoch {epoch+1} Summary | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | Val Loss: {current_val_loss:.4f}")

        if current_val_loss < best_val_loss:
            best_val_loss = current_val_loss
            torch.save(model.state_dict(), save_path)
            print(f"   🌟 New High Score! V3 Model saved with Val Loss: {best_val_loss:.4f}")
        else:
            print(f"   ❌ Overfitting detected. Keeping previous best V3 model.")

    print("\n" + "="*50)
    print("TRAINING COMPLETE. VERSION 3 SAFELY SAVED.")
    print("="*50)
```

# ONNX for Version 3

```python
"""
=============================================================================
PRODUCTION COMPILER: THUNDERBUDDIES V3 (PyTorch -> ONNX)
=============================================================================
PURPOSE:
To export our PyTorch Audio Forensics model into the universally supported 
ONNX format for ultra-fast, CPU-only production inference.

PIPELINE STEPS:
1. Architecture Recreation: Rebuilds the ResNet18 + LSTM architecture.
2. Weight Loading: Injects the trained '.pth' weights and sets eval() mode 
   to lock BatchNormalization and Dropout layers.
3. Graph Tracing: Pushes a dummy spectrogram tensor (1x1x128x80) through the 
   network so PyTorch can mathematically map the computational graph.
4. ONNX Optimization: Applies constant folding (pre-computing static math) 
   and enables dynamic batching so the production server can process multiple 
   audio chunks simultaneously.
=============================================================================
"""
import torch
import torch.nn as nn
import torchvision.models as models

# --- 1. REBUILD THE BLUEPRINTS (One last time) ---
class ThunderBuddiesCRNN(nn.Module):
    def __init__(self):
        super(ThunderBuddiesCRNN, self).__init__()
        self.resnet = models.resnet18(weights=None)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        self.features = nn.Sequential(*list(self.resnet.children())[:-1])
        self.lstm = nn.LSTM(input_size=512, hidden_size=128, num_layers=1, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(256, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.features(x)                  
        x = x.squeeze(-1).squeeze(-1)         
        x = x.unsqueeze(1)                    
        lstm_out, _ = self.lstm(x)            
        x = lstm_out[:, -1, :]                
        x = self.fc(x)
        x = self.sigmoid(x)                   
        return x

print("Loading PyTorch model into memory...")
device = torch.device("cpu")
model = ThunderBuddiesCRNN().to(device)

# Load your Version 3 weights
model.load_state_dict(torch.load("thunderbuddies_model_v3.pth", map_location=device, weights_only=True))
model.eval() # CRITICAL: Sets dropout and batchnorm to evaluation mode

# --- 2. CREATE A DUMMY AUDIO FILE ---
# This matches your exact spectrogram shape: (Batch Size 1, 1 Color Channel, 128 Frequencies, 80 Time Steps)
dummy_audio_input = torch.randn(1, 1, 128, 80, device=device)

# --- 3. EXPORT TO ONNX ---
onnx_file_path = "thunderbuddies_v3.onnx"
print("Freezing graph and exporting to ONNX...")

torch.onnx.export(
    model, 
    dummy_audio_input, 
    onnx_file_path,
    export_params=True,
    opset_version=14,          # Opset 14 is the industry standard for stable deployment
    do_constant_folding=True,  # This mathematically shrinks the model for extreme speed
    input_names=['input'],     # Name the entry gate
    output_names=['output'],   # Name the exit gate
    dynamic_axes={
        'input': {0: 'batch_size'}, 
        'output': {0: 'batch_size'}
    } # This allows you to process 1 clip or 10 clips at the same time later
)

print(f"✅ Success! Your production model is ready: {onnx_file_path}")
```

# Testing ONNX of Version 3 (looks only at the first 2.5 seconds)

```python
'''
The Restriction: It contains the line duration=2.5 when loading the audio. This means the model blindly looks at only the first 2.5 seconds of the file and ignores everything else.
No Filtering: It processes whatever is in those first 2.5 seconds, even if it is pure silence or background static.
Single Inference: It runs the neural network exactly once.
Visuals: It only generates a 2-pane dashboard (Spectrogram and Probability Bar).
'''
import os
import time
import librosa
import numpy as np
import onnxruntime as ort
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import warnings

# Mute warnings for a clean output
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# --- 1. THE ANALYSIS ENGINE ---
def analyze_static_audio(audio_path, model_path="thunderbuddies_v3.onnx"):
    print(f"Loading ONNX engine and analyzing: {os.path.basename(audio_path)}...")
    
    sess_options = ort.SessionOptions()
    sess_options.intra_op_num_threads = 32  # Optimized for your high-CPU machine
    session = ort.InferenceSession(model_path, sess_options, providers=['CPUExecutionProvider'])
    input_name = session.get_inputs()[0].name
    output_name = session.get_outputs()[0].name

    try:
        # Process Audio 
        y, sr = librosa.load(audio_path, sr=16000, duration=2.5)
        mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
        mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
        
        # Shape the Data
        spec = np.expand_dims(mel_spec_db, axis=0) 
        spec = np.expand_dims(spec, axis=0) 
        
        TARGET_WIDTH = 80
        current_width = spec.shape[3]
        
        if current_width < TARGET_WIDTH:
            pad_width = TARGET_WIDTH - current_width
            spec = np.pad(spec, ((0,0), (0,0), (0,0), (0, pad_width)), mode='constant', constant_values=-80.0)
        elif current_width > TARGET_WIDTH:
            spec = spec[:, :, :, :TARGET_WIDTH]
            
        payload = spec.astype(np.float32)

        # Run ONNX Inference
        start_time = time.time()
        prediction = session.run([output_name], {input_name: payload})[0][0][0]
        exec_time = (time.time() - start_time) * 1000

        fake_prob = prediction * 100
        real_prob = (1 - prediction) * 100

        # Print Terminal Results
        print("\n" + "="*50)
        if prediction > 0.5:
            print(f"🚨 HIGH RISK: SYNTHETIC ARTIFACTS DETECTED")
        else:
            print(f"✅ VERIFIED: AUTHENTIC HUMAN VOICE")
        print(f"   Fake Probability: {fake_prob:.2f}%")
        print(f"   Real Probability: {real_prob:.2f}%")
        print(f"⚡ ONNX Latency:     {exec_time:.2f} ms")
        print("="*50 + "\n")

        # Generate the Visual Dashboard
        fig, axes = plt.subplots(1, 2, figsize=(12, 4), gridspec_kw={'width_ratios': [2, 1]})
        
        # Pane 1: The Spectrogram
        ax1 = axes[0]
        im = ax1.imshow(mel_spec_db, aspect='auto', origin='lower', cmap='viridis')
        ax1.set_title("Mel-Spectrogram (Acoustic Fingerprint)")
        ax1.set_xlabel("Time Frames")
        ax1.set_ylabel("Frequency Bins")
        fig.colorbar(im, ax=ax1, format="%+2.0f dB")

        # Pane 2: The Probability Distribution
        ax2 = axes[1]
        labels = ['Authentic Human', 'AI Generated']
        probabilities = [real_prob, fake_prob]
        colors = ['mediumseagreen', 'crimson']
        
        bars = ax2.bar(labels, probabilities, color=colors)
        ax2.set_ylim(0, 110)
        ax2.set_ylabel("Confidence Score (%)")
        ax2.set_title("Detection Probability")
        
        for bar in bars:
            yval = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2, yval + 2, f"{yval:.1f}%", ha='center', va='bottom', fontweight='bold')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Analysis Failed: {e}")

# --- 2. THE INTERACTIVE UPLOAD UI ---
upload_button = widgets.FileUpload(
    accept='audio/*',  
    multiple=False, 
    description='Upload Audio'
)
output_area = widgets.Output()

def on_upload(change):
    # Ignore empty button clears
    if not upload_button.value:
        return
        
    with output_area:
        clear_output()
        print("Processing upload through ONNX Engine...")
        
        uploaded_file = upload_button.value
        # Handle different ipywidgets versions (tuple vs dict)
        if isinstance(uploaded_file, tuple): 
            content = uploaded_file[0]['content']
            filename = uploaded_file[0]['name']
        else: 
            filename = list(uploaded_file.keys())[0]
            content = uploaded_file[filename]['content']
            
        _, ext = os.path.splitext(filename)
        temp_path = f"live_test_clip{ext}"
        
        # Save uploaded file to disk
        with open(temp_path, "wb") as f:
            f.write(content)
            
        # Run the visual analysis!
        analyze_static_audio(temp_path)
        
        # Safely reset the button for the next file
        upload_button.unobserve(on_upload, names='value')
        if isinstance(upload_button.value, tuple):
            upload_button.value = ()
        else:
            upload_button.value.clear()
        upload_button.observe(on_upload, names='value')

# Bind the function to the button
upload_button.observe(on_upload, names='value')

# Display the UI
display(upload_button, output_area)
```

# Testing Version 3 using ONNX (looks at the whole clip and generates an additional waveform, improvement over the previous cell)

```python
'''
Full Audio Loading: The duration=2.5 limit is removed. It loads the entire file.
Sliding Window: It chops the full audio into 2.5-second chunks and processes them sequentially in a loop.
Voice Activity Detection (VAD): Before running the heavy neural network, it calculates the RMS energy (volume) of each chunk. If the chunk is silent (< 0.005), it skips it to save processing power and prevent false positives from static.
Aggregation: It runs the model multiple times (once per valid chunk) and averages the scores together for a highly accurate final verdict.
Visuals: Upgraded to a 3-pane dashboard (adding the Full Waveform)
'''
import os
import time
import librosa
import numpy as np
import onnxruntime as ort
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import warnings

# Mute warnings for a clean output
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

def analyze_full_static_audio(audio_path, model_path="thunderbuddies_v3.onnx"):
    print(f"Loading ONNX engine and analyzing Full Clip: {os.path.basename(audio_path)}...")
    
    sess_options = ort.SessionOptions()
    sess_options.intra_op_num_threads = 32  # Optimized for your high-CPU machine
    session = ort.InferenceSession(model_path, sess_options, providers=['CPUExecutionProvider'])
    input_name = session.get_inputs()[0].name
    output_name = session.get_outputs()[0].name

    try:
        start_total_time = time.time()

        # 1. Load the ENTIRE audio file (Removed the duration=2.5 limit)
        y_full, sr = librosa.load(audio_path, sr=16000)
        
        # Calculate chunk size (2.5 seconds = 40,000 samples)
        chunk_size = int(16000 * 2.5)
        deepfake_scores = []
        
        print(f"File loaded. Total duration: {len(y_full)/sr:.2f} seconds. Scanning all chunks...")

        # 2. Sliding Window Loop
        for i in range(0, len(y_full), chunk_size):
            y_chunk = y_full[i:i+chunk_size]
            
            # Skip the very last chunk if it's less than 1 second
            if len(y_chunk) < 16000 * 1.0: 
                continue
            
            # --- VOICE ACTIVITY DETECTOR (VAD) FILTER ---
            # Measure the volume/energy of the chunk to skip pure silence
            rms_energy = np.mean(librosa.feature.rms(y=y_chunk))
            if rms_energy < 0.005:
                continue
            # --------------------------------------------
                
            # Process Audio Chunk for ONNX
            mel_spec = librosa.feature.melspectrogram(y=y_chunk, sr=sr, n_mels=128, fmax=8000)
            mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
            
            spec = np.expand_dims(mel_spec_db, axis=0) 
            spec = np.expand_dims(spec, axis=0) 
            
            TARGET_WIDTH = 80
            current_width = spec.shape[3]
            
            if current_width < TARGET_WIDTH:
                pad_width = TARGET_WIDTH - current_width
                spec = np.pad(spec, ((0,0), (0,0), (0,0), (0, pad_width)), mode='constant', constant_values=-80.0)
            elif current_width > TARGET_WIDTH:
                spec = spec[:, :, :, :TARGET_WIDTH]
                
            payload = spec.astype(np.float32)

            # Run ONNX Inference for this specific chunk
            prediction = session.run([output_name], {input_name: payload})[0][0][0]
            deepfake_scores.append(prediction)

        # 3. Aggregate Results
        exec_time = (time.time() - start_total_time) * 1000
        
        # Protect against edge case where the entire clip was silent
        avg_prediction = np.mean(deepfake_scores) if deepfake_scores else 0.0
        fake_prob = avg_prediction * 100
        real_prob = (1 - avg_prediction) * 100

        # Print Terminal Results
        print("\n" + "="*50)
        if avg_prediction > 0.5:
            print(f"🚨 HIGH RISK: SYNTHETIC ARTIFACTS DETECTED")
        else:
            print(f"✅ VERIFIED: AUTHENTIC HUMAN VOICE")
        print(f"   Overall Fake Probability: {fake_prob:.2f}%")
        print(f"   Overall Real Probability: {real_prob:.2f}%")
        print(f"⚡ ONNX Latency (Full Clip): {exec_time:.2f} ms")
        print("="*50 + "\n")

        # 4. Generate the TRIPLE-PANE Visual Dashboard for the FULL audio
        mel_spec_full = librosa.feature.melspectrogram(y=y_full, sr=sr, n_mels=128, fmax=8000)
        mel_spec_full_db = librosa.power_to_db(mel_spec_full, ref=np.max)

        fig, axes = plt.subplots(1, 3, figsize=(16, 4), gridspec_kw={'width_ratios': [2, 2, 1]})
        fig.suptitle(f"Audio Forensics Analysis: {os.path.basename(audio_path)}", fontsize=14, fontweight='bold')

        # --- PANE 1: The Waveform ---
        ax1 = axes[0]
        times = np.arange(len(y_full)) / float(sr) 
        ax1.plot(times, y_full, color='dodgerblue', alpha=0.8)
        ax1.set_title("Waveform (Full Clip)")
        ax1.set_xlabel("Time (Seconds)")
        ax1.set_ylabel("Amplitude")
        ax1.set_ylim(-1.0, 1.0) 

        # --- PANE 2: The Spectrogram ---
        ax2 = axes[1]
        im = ax2.imshow(mel_spec_full_db, aspect='auto', origin='lower', cmap='viridis')
        ax2.set_title("Mel-Spectrogram (Full Clip)")
        ax2.set_xlabel("Time Frames")
        ax2.set_ylabel("Frequency Bins")
        fig.colorbar(im, ax=ax2, format="%+2.0f dB")

        # --- PANE 3: The Probability Distribution ---
        ax3 = axes[2]
        labels = ['Authentic Human', 'AI Generated']
        probabilities = [real_prob, fake_prob]
        colors = ['mediumseagreen', 'crimson']
        
        bars = ax3.bar(labels, probabilities, color=colors)
        ax3.set_ylim(0, 110)
        ax3.set_ylabel("Avg Confidence Score (%)")
        ax3.set_title("Detection Probability")
        
        for bar in bars:
            yval = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2, yval + 2, f"{yval:.1f}%", ha='center', va='bottom', fontweight='bold')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Analysis Failed: {e}")


# --- 5. THE INTERACTIVE UPLOAD UI ---
upload_button = widgets.FileUpload(
    accept='audio/*',  
    multiple=False, 
    description='Upload Audio'
)
output_area = widgets.Output()

def on_upload(change):
    # Ignore empty button clears
    if not upload_button.value:
        return
        
    with output_area:
        clear_output()
        print("Processing upload through ONNX Engine...")
        
        uploaded_file = upload_button.value
        if isinstance(uploaded_file, tuple): 
            content = uploaded_file[0]['content']
            filename = uploaded_file[0]['name']
        else: 
            filename = list(uploaded_file.keys())[0]
            content = uploaded_file[filename]['content']
            
        _, ext = os.path.splitext(filename)
        temp_path = f"live_test_clip{ext}"
        
        # Save uploaded file to disk
        with open(temp_path, "wb") as f:
            f.write(content)
            
        # Run the full-clip visual analysis!
        analyze_full_static_audio(temp_path)
        
        # Safely reset the button for the next file
        upload_button.unobserve(on_upload, names='value')
        if isinstance(upload_button.value, tuple):
            upload_button.value = ()
        else:
            upload_button.value.clear()
        upload_button.observe(on_upload, names='value')
        

# Bind the function to the button
upload_button.observe(on_upload, names='value')

# Display the UI
display(upload_button, output_area)
```

# Generation of Evaluation Metrics, Confusion Matrix and KDE Plot for Version 3

```python
import os
import glob
import numpy as np
import onnxruntime as ort
from torch.utils.data import Dataset, DataLoader, random_split
import torch
import torch.nn.functional as F
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from tqdm.notebook import tqdm
import warnings

# Mute warnings for a clean output
warnings.filterwarnings("ignore")

# --- 1. REBUILD THE DATASET PIPELINE ---
class AudioForensicsDataset(Dataset):
    def __init__(self, data_dir):
        self.files = glob.glob(os.path.join(data_dir, '**/*.npy'), recursive=True)
        
    def __len__(self):
        return len(self.files)
        
    def __getitem__(self, idx):
        file_path = self.files[idx]
        spec = np.load(file_path)
        spec = np.expand_dims(spec, axis=0) 
        spec_tensor = torch.tensor(spec, dtype=torch.float32)
        
        # Cookie Cutter
        TARGET_WIDTH = 80
        current_width = spec_tensor.shape[2]
        if current_width < TARGET_WIDTH:
            padding = TARGET_WIDTH - current_width
            spec_tensor = F.pad(spec_tensor, (0, padding))
        elif current_width > TARGET_WIDTH:
            spec_tensor = spec_tensor[:, :, :TARGET_WIDTH]
        
        label = 1.0 if 'Fake Audio' in file_path else 0.0
        return spec_tensor, torch.tensor([label], dtype=torch.float32)

# --- 2. LOAD DATA AND ONNX ENGINE ---
print("Rebuilding dataset and loading ONNX engine...")
dataset = AudioForensicsDataset('./spectrogram_data/')

train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

# Set seed for reproducible splits so you test the exact same files
generator = torch.Generator().manual_seed(42)
_, val_dataset = random_split(dataset, [train_size, val_size], generator=generator)

# DataLoader handles the heavy lifting of batching the files
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=2)

# Boot up the ONNX Model using 32 Cores
sess_options = ort.SessionOptions()
sess_options.intra_op_num_threads = 32  
session = ort.InferenceSession("thunderbuddies_v3.onnx", sess_options, providers=['CPUExecutionProvider'])
input_name = session.get_inputs()[0].name
output_name = session.get_outputs()[0].name

# --- 3. RUN INFERENCE ---
y_true = []
y_scores = []

print(f"Running ONNX inference on {val_size} validation clips. Please wait...")
for inputs, labels in tqdm(val_loader, desc="Evaluating"):
    # Convert PyTorch batched tensors to standard NumPy arrays
    inputs_np = inputs.numpy()
    
    # Bulk predict 64 clips at a time through ONNX
    outputs = session.run([output_name], {input_name: inputs_np})[0]
    
    y_scores.extend(outputs.flatten())
    y_true.extend(labels.numpy().flatten())

y_true = np.array(y_true)
y_scores = np.array(y_scores)
y_pred = (y_scores > 0.5).astype(int)

# --- 4. PRINT METRICS & PLOT ---
print("\n" + "="*40)
print("FINAL ONNX EVALUATION METRICS")
print("="*40)
print(f"Accuracy:  {accuracy_score(y_true, y_pred):.4f}")
print(f"Precision: {precision_score(y_true, y_pred):.4f}")
print(f"Recall:    {recall_score(y_true, y_pred):.4f}")
print(f"F1-Score:  {f1_score(y_true, y_pred):.4f}")
print("="*40)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0], 
            xticklabels=['Real (0)', 'Fake (1)'], 
            yticklabels=['Real (0)', 'Fake (1)'],
            annot_kws={"size": 14})
axes[0].set_title('Audio Forensics: Confusion Matrix', fontsize=14)
axes[0].set_ylabel('Actual Label', fontsize=12)
axes[0].set_xlabel('Predicted Label', fontsize=12)

# KDE Plot
sns.kdeplot(y_scores[y_true == 0], fill=True, color="mediumseagreen", label="Real Human (Actual)", ax=axes[1])
sns.kdeplot(y_scores[y_true == 1], fill=True, color="crimson", label="AI Clone (Actual)", ax=axes[1])
axes[1].set_title('KDE Plot: Model Confidence Separation', fontsize=14)
axes[1].set_xlabel('Predicted Probability (0 = Real, 1 = Fake)', fontsize=12)
axes[1].set_ylabel('Density', fontsize=12)
axes[1].legend()

plt.tight_layout()
plt.show()
```

# Replacing Mel-Spectrogram with Wav2Vec 2.0. Transitioning from image-based approach to Deep Sequence / Embedding-Based Approach

```python
"""
=============================================================================
WAV2VEC 2.0 PIPELINE - ACOUSTIC FEATURE EXTRACTION
=============================================================================
This script is a critical preprocessing module for the V6 Audio Forensics engine.
It replaces our legacy Librosa Mel-Spectrogram generator.

HOW IT WORKS "UNDER THE HOOD":
1. Chunking & VAD: Like previous versions, it slices audio into 2.5s chunks 
   and uses RMS energy to ignore pure silence.
2. The Processor: Normalizes the raw 16kHz 1D waveform into the exact tensor 
   format expected by the Wav2Vec 2.0 Transformer.
3. The Transformer: Passes the waveform through multiple self-attention layers. 
   We freeze the gradients (torch.no_grad()) because we are only USING the 
   model's pre-existing knowledge of human speech, not re-training it.
4. Temporal Pooling: The raw output shape is (Batch, TimeSteps, 768). We apply 
   mean pooling across the time dimension (`mean(dim=1)`) to collapse the time 
   steps. This compresses the entire 2.5 seconds of audio into a single, static 
   768-dimensional vector (an array of 768 numbers).

OUTPUT:
Returns a 2D NumPy matrix of shape (Number_of_Valid_Chunks, 768).
=============================================================================
"""
import os
import time
import librosa
import numpy as np
import torch
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2Model
import warnings

warnings.filterwarnings("ignore")

print("Downloading SOTA Extractor (Wav2Vec 2.0)...")

# --- 1. LOAD THE PRE-TRAINED SOTA MODEL ---
# The Feature Extractor ensures the audio is formatted perfectly for the model
processor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/wav2vec2-base")
# The Model itself (running on CPU)
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base")
model.eval() # Set to evaluation mode

# --- 2. THE EMBEDDING GENERATOR ---
def extract_embeddings(audio_path):
    print(f"\nExtracting Deep Features: {os.path.basename(audio_path)}...")
    
    try:
        start_time = time.time()
        
        # Load audio (Wav2Vec 2.0 STRICTLY requires 16,000 Hz)
        y_full, sr = librosa.load(audio_path, sr=16000)
        chunk_size = int(16000 * 2.5) # 2.5 second chunks
        
        chunk_embeddings = []
        
        for i in range(0, len(y_full), chunk_size):
            y_chunk = y_full[i:i+chunk_size]
            
            if len(y_chunk) < 16000 * 1.0: 
                continue
                
            # --- VAD FILTER ---
            rms_energy = np.mean(librosa.feature.rms(y=y_chunk))
            if rms_energy < 0.005:
                continue
                
            # ==========================================
            # NEW: SOTA EXTRACTION (Replaces Mel-Spec)
            # ==========================================
            # 1. Prepare the raw waveform for the model
            inputs = processor(y_chunk, sampling_rate=16000, return_tensors="pt")
            
            # 2. Pass it through the Transformer layers (No gradient tracking needed)
            with torch.no_grad():
                outputs = model(**inputs)
            
            # 3. outputs.last_hidden_state shape is (Batch, TimeSteps, 768)
            # We average it across the TimeSteps to get one single 768-number fingerprint for this chunk
            embedding = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
            
            chunk_embeddings.append(embedding)
            # ==========================================

        exec_time = time.time() - start_time
        
        if not chunk_embeddings:
            print("⚠️ No speech detected to extract.")
            return None
            
        # Convert list of arrays into a single 2D NumPy matrix
        final_matrix = np.vstack(chunk_embeddings)
        
        print(f"✅ Extraction Complete in {exec_time:.2f}s")
        print(f"📊 Matrix Shape: {final_matrix.shape} -> (Chunks Analyzed, Hidden Features)")
        
        return final_matrix

    except Exception as e:
        print(f"Extraction Failed: {e}")
        return None

# --- TEST THE EXTRACTOR ---
# Pass one of your audio files through it to see the new data structure
embeddings = extract_embeddings("live_test_clip.ogg")

if embeddings is not None:
    print("\nSneak peek at the first 5 numbers of Chunk 1's embedding:")
    print(embeddings[0][:5])
```

```python
"""
Multi-Vendor Deepfake Synthesis & Degradation Pipeline
------------------------------------------------------
This pipeline creates a robust 130-file test dataset to evaluate the final 
Audio Forensics Meta-Classifier against unseen, state-of-the-art TTS engines.

PIPELINE MECHANICS:
1. Data Ingestion: Reads 20 authentic sentences per language from our CSV dataset (synthetic_booster_validated_sentences.csv).
2. Vendor Routing: 
   - Sentences 1-10 are sent to Google Cloud TTS (using high-fidelity Neural2 
     and WaveNet models depending on language availability).
   - Sentences 11-20 are sent to OpenAI's 'Nova' voice (tts-1-hd model).
   - *Note: Bengali is explicitly excluded from OpenAI due to lack of support.
3. The Augmentation Engine (save_with_noise):
   To prevent our forensics model from simply learning that "no background noise 
   equals AI," we intercept the raw bytes from the APIs. We load them into an 
   array, calculate the maximum amplitude, and inject a randomized threshold 
   of white noise. We then export the "dirty" file.
4. Output: Saves all files uniformly to the './synthetic_booster_pack' directory.
------------------------------------------------------
"""
import os
import time
import pandas as pd
import numpy as np
import librosa
import soundfile as sf
from openai import OpenAI
from google.cloud import texttospeech
from tqdm.notebook import tqdm
import warnings

warnings.filterwarnings("ignore")

# --- 1. SETUP & AUTHENTICATION ---
# Set your OpenAI API Key here
os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_API_KEY" 
client = OpenAI()
gcp_client = texttospeech.TextToSpeechClient()

output_dir = "./synthetic_booster_pack"
os.makedirs(output_dir, exist_ok=True)

# --- 2. LANGUAGE MAPPING & RULES ---
# 1. Map the short codes to the Google Cloud voices
gcp_voice_map = {
    'en': 'en-US-Journey-D',      # Journey (SOTA)
    'fr': 'fr-FR-Neural2-A',      # Neural2 (Ultra-high quality)
    'es': 'es-ES-Neural2-A',
    'de': 'de-DE-Neural2-B',
    'zh-CN': 'cmn-CN-Wavenet-A',  # Wavenet
    'ca': 'ca-ES-Standard-A',     # Standard
    'bn': 'bn-IN-Wavenet-A'
}

# 2. Map your CSV's exact words to the API's required short codes
csv_to_2letter = {
    'English': 'en',
    'French': 'fr',
    'Spanish': 'es',
    'German': 'de',
    'Chinese': 'zh-CN',
    'Catalan': 'ca',
    'Bengali': 'bn'
}

# The languages OpenAI supports
openai_languages = ['en', 'fr', 'es', 'de', 'zh-CN', 'ca'] # No Bengali

# --- 3. LOAD DATASET ---
print("Loading dataset...")
df = pd.read_csv("synthetic_booster_validated_sentences.csv")
col_lang = 'Language'     
col_text = 'Sentence'   

# --- 4. THE AUGMENTATION ENGINE ---
def save_with_noise(raw_audio_content, filename):
    """Saves raw API bytes to disk, adds white noise, and resaves as a clean .wav"""
    temp_path = f"temp_{filename}"
    final_path = os.path.join(output_dir, filename)
    
    with open(temp_path, "wb") as out:
        out.write(raw_audio_content)
        
    y, sr = librosa.load(temp_path, sr=16000)
    
    # Inject microscopic white noise (The "Dirty Room" trick)
    noise_amp = 0.005 * np.random.uniform() * np.amax(y)
    y_noisy = y + noise_amp * np.random.normal(size=y.shape[0])
    
    sf.write(final_path, y_noisy, sr)
    os.remove(temp_path)

# --- 5. API GENERATORS ---
def generate_gcp(text, lang_code, index):
    try:
        synthesis_input = texttospeech.SynthesisInput(text=text)
        voice_name = gcp_voice_map.get(lang_code, 'en-US-Standard-A')
        
        voice = texttospeech.VoiceSelectionParams(
            language_code=voice_name[:5], 
            name=voice_name 
        )
        audio_config = texttospeech.AudioConfig(
            audio_encoding=texttospeech.AudioEncoding.LINEAR16,
            sample_rate_hertz=16000
        )
        response = gcp_client.synthesize_speech(
            input=synthesis_input, voice=voice, audio_config=audio_config
        )
        filename = f"gcp_{lang_code}_{index:02d}.wav"
        save_with_noise(response.audio_content, filename)
        return True
    except Exception as e:
        print(f"❌ GCP Error on {lang_code}_{index}: {e}")
        return False

def generate_openai(text, lang_code, index):
    try:
        response = client.audio.speech.create(
            model="tts-1-hd", 
            voice="nova",     
            input=text
        )
        filename = f"openai_{lang_code}_{index:02d}.wav"
        save_with_noise(response.content, filename)
        return True
    except Exception as e:
        print(f"❌ OpenAI Error on {lang_code}_{index}: {e}")
        return False

# --- 6. THE SMARTER MASS GENERATION LOOP ---
print("\n🚀 Starting SOTA Factory...")
total_gcp = 0
total_openai = 0

for csv_lang, strict_lang_code in csv_to_2letter.items():
    
    # Filter using the exact word found in your CSV (e.g., 'English')
    lang_df = df[df[col_lang] == csv_lang].head(20)
    sentences = lang_df[col_text].tolist()
    
    print(f"\nProcessing Language: [{csv_lang.upper()}] - {len(sentences)} sentences found.")
    
    for i, text in enumerate(sentences):
        # Sentences 0 to 9 -> Google Cloud TTS
        if i < 10:
            success = generate_gcp(text, strict_lang_code, i)
            if success: total_gcp += 1
            time.sleep(0.5) # Prevent GCP rate limits
            
        # Sentences 10 to 19 -> OpenAI TTS (Skip if Bengali)
        elif i >= 10 and strict_lang_code in openai_languages:
            success = generate_openai(text, strict_lang_code, i)
            if success: total_openai += 1
            time.sleep(1) # Be polite to OpenAI rate limits

print("\n" + "="*40)
print("🎉 FACTORY RUN COMPLETE")
print("="*40)
print(f"Google Cloud Clips Generated : {total_gcp} / 70")
print(f"OpenAI Clips Generated       : {total_openai} / 60")
print(f"Total Deepfakes Ready        : {total_gcp + total_openai} / 130")
print("All files saved to './synthetic_booster_pack' with environmental noise injected.")
```

# Model Version 4 Training

```python
"""
=============================================================================
AUDIO FORENSICS V4.0 - STANDARD TRAINING PIPELINE
=============================================================================
PURPOSE:
This script represents the 'Training Phase' of the machine learning lifecycle. 
It processes thousands of Authentic, Legacy Fake, and SOTA Booster audio clips 
to extract 768-D Wav2Vec 2.0 embeddings. It then uses these embeddings to train 
the XGBoost Meta-Classifier and saves the final learned weights to disk as
'thunderbuddies_v4.json'.
=============================================================================
"""
import os
import glob
import time
import librosa
import numpy as np
import torch
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2Model
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import warnings

warnings.filterwarnings("ignore")

# ==========================================
# 1. PIPELINE CONFIGURATION
# ==========================================
LOCAL_DATA_DIR = "./local_data"
BOOSTER_PACK_DIR = "./synthetic_booster_pack"

print("Booting up V100 GPU for Master Production Run...")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Active Deep Learning Device: {device.type.upper()}")

# ==========================================
# 2. LOAD SOTA EXTRACTOR (Wav2Vec 2.0)
# ==========================================
processor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/wav2vec2-base")
extractor_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base").to(device)
extractor_model.eval()

# ==========================================
# 3. THE FEATURE EXTRACTION ENGINE
# ==========================================
def extract_features(audio_path):
    """Loads audio, applies VAD silence filtering, and extracts 768-D embeddings."""
    try:
        y_full, sr = librosa.load(audio_path, sr=16000)
        chunk_size = int(16000 * 2.5)
        chunk_embeddings = []
        
        for i in range(0, len(y_full), chunk_size):
            y_chunk = y_full[i:i+chunk_size]
            
            if len(y_chunk) < 16000 * 1.0: 
                continue
                
            # VAD Filter: Skip dead air
            if np.mean(librosa.feature.rms(y=y_chunk)) < 0.005: 
                continue
                
            # Move chunk to GPU and extract
            inputs = processor(y_chunk, sampling_rate=16000, return_tensors="pt")
            inputs = {k: v.to(device) for k, v in inputs.items()}
            
            with torch.no_grad():
                outputs = extractor_model(**inputs)
            
            # Average the time steps into a single 768-D vector
            embedding = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
            chunk_embeddings.append(embedding)
            
        if not chunk_embeddings: 
            return None
        
        # Average all spoken chunks to get ONE master fingerprint for the entire clip
        master_embedding = np.mean(np.vstack(chunk_embeddings), axis=0)
        return master_embedding

    except Exception as e:
        # Failsafe: If an audio file is corrupted, skip it without crashing the 10-hour run
        return None

# ==========================================
# 4. MASS DATA HARVESTING 
# ==========================================
print("\nScanning local storage for audio assets...")
real_files = glob.glob(os.path.join(LOCAL_DATA_DIR, "*", "*Real Audio*", "*.wav"))
fake_legacy_files = glob.glob(os.path.join(LOCAL_DATA_DIR, "*", "*Fake Audio*", "*.wav"))
fake_sota_files = glob.glob(os.path.join(BOOSTER_PACK_DIR, "*.wav"))

print(f" - Authentic Human Clips: {len(real_files)}")
print(f" - Legacy Fake Clips:     {len(fake_legacy_files)}")
print(f" - SOTA Booster Clips:    {len(fake_sota_files)}")
print("Beginning massive parallel extraction. This will take time...\n")

X_data = []
y_labels = []

# A. Process REAL Audio (Label = 0)
for f in tqdm(real_files, desc="Extracting Real Human Audio"):
    features = extract_features(f)
    if features is not None:
        X_data.append(features)
        y_labels.append(0)

# B. Process LEGACY FAKE Audio (Label = 1)
for f in tqdm(fake_legacy_files, desc="Extracting Legacy Deepfakes"):
    features = extract_features(f)
    if features is not None:
        X_data.append(features)
        y_labels.append(1)

# C. Process SOTA BOOSTER PACK (Label = 1)
# We append these to the same Fake class to ensure the model catches modern threats
for f in tqdm(fake_sota_files, desc="Extracting SOTA Booster Pack"):
    features = extract_features(f)
    if features is not None:
        X_data.append(features)
        y_labels.append(1)

X = np.array(X_data)
y = np.array(y_labels)
print(f"\n✅ Final Matrix Built: {X.shape[0]} clips vectorized into {X.shape[1]} acoustic features.")

# ==========================================
# 5. HIGH-FIDELITY XGBOOST TRAINING
# ==========================================
print("\nInitiating Maximum-Quality XGBoost Training Sequence on V100 GPU...")

# We use an 80/10/10 split (Train/Validation/Test)
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.11, stratify=y_temp, random_state=42)

# Enterprise Hyperparameters
xgb_model = xgb.XGBClassifier(
    n_estimators=2000,         # Max out the trees, Early Stopping will cut it off when perfect
    max_depth=7,               # Deeper trees for complex audio patterns
    learning_rate=0.01,        # Slower learning rate for maximum precision
    tree_method='hist',        # GPU optimized
    device='cuda',             # FORCE XGBoost to use your NVIDIA V100
    scale_pos_weight=1.2,      # Slight bias to heavily penalize False Negatives
    early_stopping_rounds=50,  # Prevent overfitting
    eval_metric='auc',         # Optimize for Area Under the ROC Curve
    random_state=42
)

start_train = time.time()
# Train the model, using the Validation set to test it after every tree is built
xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=100  # Print an update every 100 trees
)
train_time = time.time() - start_train
print(f"\n✅ Neural Weights Optimized in {train_time:.2f} seconds.")

model_filename = "thunderbuddies_v4.json"
xgb_model.save_model(model_filename)
print(f"💾 Production Model secured at: {model_filename}")

# ==========================================
# 6. SECURITY THREAT MATRIX EVALUATION
# ==========================================
# Test the model on the final 10% of data it has NEVER seen before
y_pred_probs = xgb_model.predict_proba(X_test)[:, 1] 
y_pred = (y_pred_probs > 0.5).astype(int)

print("\n" + "="*50)
print("VERSION 4 MAXIMUM QUALITY EVALUATION METRICS")
print("="*50)
print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred):.4f}  <-- (Crucial target: >0.95)")
print(f"F1-Score:  {f1_score(y_test, y_pred):.4f}")
print("="*50)

# Visual Dashboard
plt.figure(figsize=(7, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Authentic Human (0)', 'Synthetic Audio (1)'], 
            yticklabels=['Authentic Human (0)', 'Synthetic Audio (1)'],
            annot_kws={"size": 14})
plt.title('V4 Ultimate Threat Matrix Analysis', fontsize=15, fontweight='bold')
plt.ylabel('Actual Origin', fontsize=12)
plt.xlabel('AI Prediction', fontsize=12)
plt.tight_layout()
plt.show()
```

# Model Version 4 Training, which is much safer

```python
"""
=============================================================================
AUDIO FORENSICS V4.1 - BULLETPROOF TRAINING PIPELINE (WATCHDOG ENABLED)
=============================================================================
PURPOSE:
An enterprise-grade upgrade to the V4 Training Pipeline. By combining 'torchaudio' 
with an OS-level POSIX signal watchdog (SIGALRM), this script guarantees that 
C-library decoding deadlocks caused by corrupted audio files will be forcefully 
skipped. This ensures massive, multi-hour dataset extraction runs can execute 
overnight without silently freezing the VM.
It replaces librosa with PyTorch's native torchaudio
=============================================================================
"""
import os
import glob
import time
import librosa
import numpy as np
import torch
import torchaudio
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2Model
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import warnings
import signal

warnings.filterwarnings("ignore")

# ==========================================
# 0. THE KERNEL WATCHDOG (Deadlock Protection)
# ==========================================
class TimeoutException(Exception):
    pass

def timeout_handler(signum, frame):
    raise TimeoutException("Hardware Deadlock Detected")

# Bind the alarm signal to our custom handler
signal.signal(signal.SIGALRM, timeout_handler)

# ==========================================
# 1. PIPELINE CONFIGURATION
# ==========================================
LOCAL_DATA_DIR = "./local_data"
BOOSTER_PACK_DIR = "./synthetic_booster_pack"

print("Booting up V100 GPU for Master Production Run...")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Active Deep Learning Device: {device.type.upper()}")

# ==========================================
# 2. LOAD SOTA EXTRACTOR (Wav2Vec 2.0)
# ==========================================
processor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/wav2vec2-base")
extractor_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base").to(device)
extractor_model.eval()

# ==========================================
# 3. BULLETPROOF FEATURE EXTRACTION 
# ==========================================
def extract_features(audio_path):
    """Loads audio safely, applies VAD silence filtering, and extracts 768-D embeddings."""
    try:
        # Start a 4-second countdown. If the C-library hangs, the OS will kill it.
        signal.alarm(4) 
        
        waveform, sr = torchaudio.load(audio_path)
        
        # Convert to Mono if Stereo
        if waveform.shape[0] > 1:
            waveform = torch.mean(waveform, dim=0, keepdim=True)
            
        # Resample to 16kHz for Wav2Vec
        if sr != 16000:
            resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)
            waveform = resampler(waveform)
            
        y_full = waveform.squeeze().numpy()
        chunk_size = int(16000 * 2.5)
        chunk_embeddings = []
        
        for i in range(0, len(y_full), chunk_size):
            y_chunk = y_full[i:i+chunk_size]
            
            if len(y_chunk) < 16000 * 1.0: 
                continue
                
            # VAD Filter: Skip dead air
            if np.mean(librosa.feature.rms(y=y_chunk)) < 0.005: 
                continue
                
            # Move chunk to GPU and extract
            inputs = processor(y_chunk, sampling_rate=16000, return_tensors="pt")
            inputs = {k: v.to(device) for k, v in inputs.items()}
            
            with torch.no_grad():
                outputs = extractor_model(**inputs)
            
            embedding = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
            chunk_embeddings.append(embedding)
            
        # Success! Turn off the alarm.
        signal.alarm(0) 
        
        if not chunk_embeddings: 
            return None
        
        master_embedding = np.mean(np.vstack(chunk_embeddings), axis=0)
        return master_embedding

    except TimeoutException:
        # The OS caught a freeze and aborted the file
        return "CORRUPTED"
    except Exception as e:
        # Turn off the alarm if a normal Python error occurs
        signal.alarm(0) 
        return "CORRUPTED"
    finally:
        # Absolute failsafe: guarantee the timer is reset for the next file
        signal.alarm(0)

# ==========================================
# 4. MASS DATA HARVESTING 
# ==========================================
print("\nScanning local storage for audio assets...")
real_files = glob.glob(os.path.join(LOCAL_DATA_DIR, "*", "*Real Audio*", "*.wav"))
fake_legacy_files = glob.glob(os.path.join(LOCAL_DATA_DIR, "*", "*Fake Audio*", "*.wav"))
fake_sota_files = glob.glob(os.path.join(BOOSTER_PACK_DIR, "*.wav"))

print(f" - Authentic Human Clips: {len(real_files)}")
print(f" - Legacy Fake Clips:     {len(fake_legacy_files)}")
print(f" - SOTA Booster Clips:    {len(fake_sota_files)}")
print("Beginning massive parallel extraction. Deadlocks will be shattered via SIGALRM...\n")

X_data = []
y_labels = []
skipped_files = 0

def process_file_list(file_list, label, desc):
    global skipped_files
    for f in tqdm(file_list, desc=desc):
        features = extract_features(f)
        if features == "CORRUPTED":
            skipped_files += 1
        elif features is not None:
            X_data.append(features)
            y_labels.append(label)

# Ingest all data
process_file_list(real_files, 0, "Extracting Real Human Audio")
process_file_list(fake_legacy_files, 1, "Extracting Legacy Deepfakes")
process_file_list(fake_sota_files, 1, "Extracting SOTA Booster Pack")

X = np.array(X_data)
y = np.array(y_labels)

print(f"\n✅ Extraction Complete!")
print(f"⚠️ Deadlocks/Corrupted Files Safely Skipped: {skipped_files}")
print(f"📊 Final Matrix Built: {X.shape[0]} valid clips vectorized into {X.shape[1]} features.")

# ==========================================
# 5. HIGH-FIDELITY XGBOOST TRAINING
# ==========================================
print("\nInitiating Maximum-Quality XGBoost Training Sequence on V100 GPU...")

X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.10, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.11, stratify=y_temp, random_state=42)

xgb_model = xgb.XGBClassifier(
    n_estimators=2000,         
    max_depth=7,               
    learning_rate=0.01,        
    tree_method='hist',        
    device='cuda',             
    scale_pos_weight=1.2,      
    early_stopping_rounds=50,  
    eval_metric='auc',         
    random_state=42
)

start_train = time.time()
xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=100  
)
train_time = time.time() - start_train
print(f"\n✅ Neural Weights Optimized in {train_time:.2f} seconds.")

model_filename = "thunderbuddies_v4.json"
xgb_model.save_model(model_filename)
print(f"💾 Production Model secured at: {model_filename}")

# ==========================================
# 6. SECURITY THREAT MATRIX EVALUATION
# ==========================================
y_pred_probs = xgb_model.predict_proba(X_test)[:, 1] 
y_pred = (y_pred_probs > 0.5).astype(int)

print("\n" + "="*50)
print("VERSION 4 MAXIMUM QUALITY EVALUATION METRICS")
print("="*50)
print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred):.4f}  <-- (Crucial target: >0.95)")
print(f"F1-Score:  {f1_score(y_test, y_pred):.4f}")
print("="*50)

plt.figure(figsize=(7, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Authentic Human (0)', 'Synthetic Audio (1)'], 
            yticklabels=['Authentic Human (0)', 'Synthetic Audio (1)'],
            annot_kws={"size": 14})
plt.title('V4 Ultimate Threat Matrix Analysis', fontsize=15, fontweight='bold')
plt.ylabel('Actual Origin', fontsize=12)
plt.xlabel('AI Prediction', fontsize=12)
plt.tight_layout()
plt.show()
```

# Version 4 Testing

```python
"""
=============================================================================
AUDIO FORENSICS V4 - INTERACTIVE LIVE INFERENCE DASHBOARD
=============================================================================
PURPOSE:
This script is the 'Deployment/Inference Phase' of the pipeline. It does NOT 
train the model. Instead, it loads the pre-trained XGBoost weights 
('thunderbuddies_v4.json') into memory and pairs it with a Jupyter-based UI.

KEY FEATURES:
1. Multi-Modal Analysis: Runs the OpenAI Whisper model (for language detection) 
   in parallel with the Wav2Vec2+XGBoost model (for deepfake detection).
2. The 5-Tier Threat Matrix: Converts raw algorithmic probabilities into 
   color-coded, actionable security alerts (e.g., '🟢 Verified' to '🛑 Critical').
3. Interactive UI: Utilizes 'ipywidgets' to allow end-users to upload wild 
   audio files and receive a visual forensic dashboard in real-time.
=============================================================================
"""
import os
import sys
import time
import torch
import librosa
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import warnings
import whisper
from transformers import Wav2Vec2Processor, Wav2Vec2Model

# Mute warnings for a clean output
warnings.filterwarnings("ignore")

# V100 BUG FIX: Disable cuDNN to prevent Conv1D engine crashes during live inference
torch.backends.cudnn.enabled = False

print("Booting up V4 Enterprise Security Engine (Wav2Vec2 + XGBoost). Please wait...")

# --- 1. GLOBALLY LOAD ALL MODELS ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load V4 Transformer Extractor
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
w2v_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base").to(device)
w2v_model.eval()

# Load V4 XGBoost Classifier
xgb_clf = xgb.XGBClassifier()
xgb_clf.load_model("thunderbuddies_v4.json")

# Load Whisper (For Language Detection)
whisper_model = whisper.load_model("tiny").to(device)

print("[*] All Models Loaded Successfully. Ready for Audio Uploads.")

# --- 2. THE 5-TIER ENTERPRISE PIPELINE ---
def analyze_full_audio(audio_path):
    print(f"Analyzing Full Clip: {os.path.basename(audio_path)}...")
    
    try:
        start_total_time = time.time()

        # Load the ENTIRE audio file
        y_full, sr = librosa.load(audio_path, sr=16000)
        
        # Calculate chunk size (2.5 seconds = 40,000 samples)
        chunk_size = int(16000 * 2.5)
        
        detected_languages = set()
        deepfake_scores = []
        
        print(f"File loaded. Total duration: {len(y_full)/sr:.2f} seconds. Scanning all chunks...")

        for i in range(0, len(y_full), chunk_size):
            y_chunk = y_full[i:i+chunk_size]
            
            if len(y_chunk) < 16000 * 1.0: 
                continue
                
            # --- VOICE ACTIVITY DETECTOR (VAD) FILTER ---
            rms_energy = np.mean(librosa.feature.rms(y=y_chunk))
            if rms_energy < 0.005:
                continue
            # --------------------------------------------
                
            # A. WHISPER: Unrestricted Language Detection
            audio_w = whisper.pad_or_trim(y_chunk.astype(np.float32))
            mel_w = whisper.log_mel_spectrogram(audio_w).to(device)
            _, probs = whisper_model.detect_language(mel_w)
            detected_code = max(probs, key=probs.get)
            
            final_lang = whisper.tokenizer.LANGUAGES.get(detected_code, detected_code).title()
            detected_languages.add(final_lang)
            
            # B. THUNDERBUDDIES V4: Deepfake Detection (Wav2Vec + XGBoost)
            trimmed_chunk, _ = librosa.effects.trim(y_chunk, top_db=20)
            
            # V4 Safety Net: Skip micro-audio
            if len(trimmed_chunk) < 8000:
                continue
                
            # Extract 768-D SOTA Embeddings
            inputs = processor(trimmed_chunk, sampling_rate=16000, return_tensors="pt")
            inputs = {k: v.to(device) for k, v in inputs.items()}
            
            with torch.no_grad():
                outputs = w2v_model(**inputs)
                
            embedding = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
            
            # XGBoost Inference (predict_proba returns [prob_real, prob_fake])
            # We reshape to (1, 768) because XGBoost expects a 2D matrix
            pred_probs = xgb_clf.predict_proba(embedding.reshape(1, -1))[0]
            deepfake_scores.append(pred_probs[1]) # Append the probability of it being Fake (1)

        # --- AGGREGATE RESULTS ---
        exec_time = (time.time() - start_total_time) * 1000
        languages_str = ", ".join(list(detected_languages)) if detected_languages else "Unknown"
        
        avg_prediction = np.mean(deepfake_scores) if deepfake_scores else 0.0
        fake_prob = avg_prediction * 100
        real_prob = (1 - avg_prediction) * 100

        # ==========================================
        # 5-TIER "HIGH RECALL" THREAT MATRIX
        # ==========================================
        if fake_prob < 5.0:
            status_icon = "🟢"
            status_text = "VERIFIED: Authentic Human"
            theme_color = "mediumseagreen" 
        elif fake_prob < 20.0:
            status_icon = "🟡"
            status_text = "LOW RISK: Minor Anomalies (Likely Real)"
            theme_color = "goldenrod"       
        elif fake_prob < 60.0:
            status_icon = "🟠"
            status_text = "ELEVATED RISK: Suspicious Audio (Review Required)"
            theme_color = "darkorange"      
        elif fake_prob < 90.0:
            status_icon = "🔴"
            status_text = "HIGH RISK: Probable AI Generation"
            theme_color = "crimson"         
        else:
            status_icon = "🛑"
            status_text = "CRITICAL: Confirmed Synthetic Deepfake"
            theme_color = "darkred"         
        # ==========================================

        # --- TERMINAL REPORT ---
        print("\n" + "="*55)
        print(f"🌍 LANGUAGES DETECTED: {languages_str.upper()}")
        print("-" * 55)
        print(f"{status_icon} {status_text}")
        print(f"   Fake Probability: {fake_prob:.2f}%")
        print(f"   Real Probability: {real_prob:.2f}%")
        print(f"⚡ Total Latency:    {exec_time:.2f} ms")
        print("="*55 + "\n")

        # --- VISUAL DASHBOARD ---
        mel_spec_full = librosa.feature.melspectrogram(y=y_full, sr=sr, n_mels=128, fmax=8000)
        mel_spec_full_db = librosa.power_to_db(mel_spec_full, ref=np.max)

        fig, axes = plt.subplots(1, 3, figsize=(16, 4), gridspec_kw={'width_ratios': [2, 2, 1]})
        
        fig.suptitle(f"Audio Forensics | {status_icon} {status_text} | File: {os.path.basename(audio_path)}", 
                     fontsize=15, fontweight='bold', color=theme_color)

        # 1. Waveform Plot
        ax1 = axes[0]
        times = np.arange(len(y_full)) / float(sr) 
        ax1.plot(times, y_full, color='dodgerblue', alpha=0.8)
        ax1.set_title("Waveform (Full Clip)")
        ax1.set_xlabel("Time (Seconds)")
        ax1.set_ylabel("Amplitude")
        ax1.set_ylim(-1.0, 1.0) 

        # 2. Spectrogram Plot (Kept for visual reference, even though V4 doesn't train on it)
        ax2 = axes[1]
        im = ax2.imshow(mel_spec_full_db, aspect='auto', origin='lower', cmap='viridis')
        ax2.set_title("Mel-Spectrogram (Visual Reference)")
        ax2.set_xlabel("Time Frames")
        ax2.set_ylabel("Frequency Bins")
        fig.colorbar(im, ax=ax2, format="%+2.0f dB")

        # 3. Probability Bar Chart
        ax3 = axes[2]
        labels = ['Authentic', 'AI Generated']
        probabilities = [real_prob, fake_prob]
        bar_colors = ['mediumseagreen', 'crimson'] 
        
        bars = ax3.bar(labels, probabilities, color=bar_colors)
        ax3.set_ylim(0, 110)
        ax3.set_ylabel("Avg Confidence Score (%)")
        ax3.set_title(status_text, color=theme_color, fontsize=10, fontweight='bold')
        
        for bar in bars:
            yval = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2, yval + 2, f"{yval:.1f}%", ha='center', va='bottom', fontweight='bold')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Analysis Failed: {e}")

# --- 3. THE INTERACTIVE UPLOAD UI ---
upload_button = widgets.FileUpload(accept='audio/*', multiple=False, description='Upload Audio')
output_area = widgets.Output()

def on_upload(change):
    if not upload_button.value:
        return
    with output_area:
        clear_output()
        
        uploaded_file = upload_button.value
        if isinstance(uploaded_file, tuple): 
            content = uploaded_file[0]['content']
            filename = uploaded_file[0]['name']
        else: 
            filename = list(uploaded_file.keys())[0]
            content = uploaded_file[filename]['content']
            
        _, ext = os.path.splitext(filename)
        temp_path = f"live_test_clip{ext}"
        
        with open(temp_path, "wb") as f:
            f.write(content)
            
        analyze_full_audio(temp_path)
        
        # Reset button state for next upload
        upload_button.unobserve(on_upload, names='value')
        if isinstance(upload_button.value, tuple):
            upload_button.value = ()
        else:
            upload_button.value.clear()
        upload_button.observe(on_upload, names='value')

upload_button.observe(on_upload, names='value')
display(upload_button, output_area)
```

# Wav2Vec 2.0 ONNX Export Protocol

```python
"""
This is a critical infrastructure script for the Audio Forensics V4 pipeline. 
Standard PyTorch models are excellent for training, but they carry massive 
library overhead during live production inference. 

PIPELINE MECHANICS:
1. We utilize HuggingFace's `Optimum` library, which contains specialized 
   C++ tracing patches specifically designed for Transformer architectures.
2. The `export=True` flag dynamically freezes the Wav2Vec 2.0 base weights 
   and translates the complex self-attention math into a static ONNX graph.
3. The resulting standalone `model.onnx` file can be executed blazingly 
   fast on standard CPUs, allowing us to hit our strict sub-10-second SLA 
   without requiring expensive GPU servers for deployment.
"""
from optimum.onnxruntime import ORTModelForFeatureExtraction

print("[*] Initializing HuggingFace Optimum Export Protocol...")

# The 'export=True' flag tells Optimum to download the base model, 
# apply the tracing patches, and compile it directly to ONNX.
ort_model = ORTModelForFeatureExtraction.from_pretrained(
    "facebook/wav2vec2-base", 
    export=True
)

# Save the finalized ONNX graph and its configuration to a dedicated folder
save_dir = "wav2vec2_onnx_production"
ort_model.save_pretrained(save_dir)

print(f"[*] SUCCESS: Enterprise ONNX model safely exported to: {save_dir}/model.onnx")
```

```python
import os
import glob
import random
import tempfile
import torch
import torch.nn as nn
import torchaudio
import numpy as np
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
from transformers import Wav2Vec2FeatureExtractor, WavLMModel
import warnings

warnings.filterwarnings("ignore")

# ==========================================
# 1. DEFINE ARCHITECTURE (Must match training)
# ==========================================
class SiameseWavLM(nn.Module):
    def __init__(self):
        super().__init__()
        self.wavlm = WavLMModel.from_pretrained("microsoft/wavlm-base-plus")
        self.classifier = nn.Sequential(
            nn.Linear(768, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 1)
        )

    def forward(self, x1, x2):
        emb1 = self.wavlm(x1).last_hidden_state.mean(dim=1)
        emb2 = self.wavlm(x2).last_hidden_state.mean(dim=1)
        diff = torch.abs(emb1 - emb2)
        return self.classifier(diff)

# ==========================================
# 2. INITIALIZE MODEL & ANCHORS
# ==========================================
print("⏳ Loading Siamese WavLM Weights...")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = SiameseWavLM().to(device)
# Load the weights you just trained
model.load_state_dict(torch.load("wavlm_siamese_v7.pth", map_location=device))
model.eval()

processor = Wav2Vec2FeatureExtractor.from_pretrained("microsoft/wavlm-base-plus")

# Grab 3 random Authentic Human clips to act as our biological baseline
real_audio_paths = glob.glob("local_data/**/Real Audio/*.wav", recursive=True)
ANCHOR_PATHS = random.sample(real_audio_paths, 3)

def load_and_pad(path, target_sr=16000, max_duration=4.0):
    waveform, sr = torchaudio.load(path, backend="soundfile")
    if sr != target_sr:
        waveform = torchaudio.transforms.Resample(sr, target_sr)(waveform)
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)
        
    waveform = waveform.squeeze().numpy()
    max_length = int(target_sr * max_duration)
    
    if len(waveform) > max_length:
        waveform = waveform[:max_length]
    else:
        waveform = np.pad(waveform, (0, max_length - len(waveform)), "constant")
        
    inputs = processor(waveform, sampling_rate=target_sr, return_tensors="pt").input_values
    return inputs

# Pre-load anchors to save time during inference
anchor_tensors = [load_and_pad(p).to(device) for p in ANCHOR_PATHS]
print("✅ Model and Anchors Ready!")

# ==========================================
# 3. INFERENCE LOGIC & UI WIDGET
# ==========================================
output_area = widgets.Output()

def analyze_audio(change):
    with output_area:
        clear_output(wait=True)
        display(HTML("<b><span style='color: orange;'>Analyzing acoustic signature...</span></b>"))
        
        # Handle ipywidgets v7 and v8 differences
        uploaded_data = change['new']
        content = None
        if isinstance(uploaded_data, tuple):  # ipywidgets >= 8.0
            content = uploaded_data[0]['content']
        elif isinstance(uploaded_data, dict): # ipywidgets < 8.0
            key = list(uploaded_data.keys())[0]
            content = uploaded_data[key]['content']
            
        if not content:
            display(HTML("<b><span style='color: red;'>Upload failed.</span></b>"))
            return

        # Write memory stream to a temporary disk file so torchaudio can safely read it
        with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as temp_wav:
            temp_wav.write(content)
            temp_path = temp_wav.name

        try:
            # 1. Load the ENTIRE audio file without max_duration limits
            waveform, sr = torchaudio.load(temp_path, backend="soundfile")
            if sr != 16000:
                waveform = torchaudio.transforms.Resample(sr, 16000)(waveform)
            if waveform.shape[0] > 1:
                waveform = torch.mean(waveform, dim=0, keepdim=True)
            waveform = waveform.squeeze().numpy()

            # 2. Define chunk parameters
            chunk_length = 16000 * 4  # 4 seconds
            stride = 16000 * 2        # 2 seconds overlap
            
            # If the audio is shorter than 4 seconds, just pad it once
            if len(waveform) <= chunk_length:
                chunks = [np.pad(waveform, (0, chunk_length - len(waveform)), "constant")]
            else:
                # Slice the audio into overlapping 4-second chunks
                chunks = [waveform[i:i + chunk_length] for i in range(0, len(waveform) - chunk_length + 1, stride)]
                # Ensure the very last piece is included and padded if necessary
                if len(waveform) % stride != 0 and len(waveform) > chunk_length:
                    last_chunk = waveform[-chunk_length:]
                    chunks.append(last_chunk)

            overall_threat_scores = []

            # 3. Analyze every chunk independently
            with torch.no_grad():
                for chunk in chunks:
                    test_tensor = processor(chunk, sampling_rate=16000, return_tensors="pt").input_values.to(device)
                    
                    chunk_threats = []
                    # Compare this chunk against all 3 anchors
                    for anchor in anchor_tensors:
                        logits = model(anchor, test_tensor)
                        prob_similar = torch.sigmoid(logits).item() 
                        threat_prob = 1.0 - prob_similar
                        chunk_threats.append(threat_prob)
                    
                    # Average the threat of this specific chunk
                    overall_threat_scores.append(sum(chunk_threats) / len(chunk_threats))
            
            # 4. Final Threat is the average threat across the ENTIRE audio file
            final_threat = (sum(overall_threat_scores) / len(overall_threat_scores)) * 100
            
            # -----------------------------------------------------
            # THE MISSING UI CODE: Display the final result!
            # -----------------------------------------------------
            clear_output(wait=True)
            if final_threat > 50.0:
                color = "crimson"
                status = "DETECTED: SYNTHETIC AUDIO"
            else:
                color = "mediumseagreen"
                status = "VERIFIED: AUTHENTIC HUMAN"
                
            display(HTML(f"""
            <div style="border: 2px solid {color}; padding: 15px; border-radius: 8px; max-width: 400px;">
                <h3 style="margin-top: 0; color: {color};">{status}</h3>
                <p style="font-size: 16px; margin-bottom: 0;">
                    <b>Deepfake Threat Score:</b> {final_threat:.2f}%
                </p>
            </div>
            """))
            # -----------------------------------------------------
            
        except Exception as e:
            display(HTML(f"<b><span style='color: red;'>Error processing audio: {e}</span></b>"))
        finally:
            os.remove(temp_path)

uploader = widgets.FileUpload(accept='.wav', multiple=False, description="Upload .wav")
uploader.observe(analyze_audio, names='value')

display(HTML("<h3>V7 Siamese Network Inference</h3>"))
display(uploader, output_area)
```

```python
import os
import torch
import torchaudio
import numpy as np
import xgboost as xgb
from transformers import Wav2Vec2FeatureExtractor, WavLMModel
import ipywidgets as widgets
from IPython.display import display, HTML
import warnings

warnings.filterwarnings("ignore")

# ==========================================
# 1. SETUP ENVIRONMENT & LOAD PATHS
# ==========================================
MODEL_PATH = "v8_production/v8_xgboost_wavlm.json"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"❌ Could not find trained model at '{MODEL_PATH}'. Check your path.")

print("⏳ Initializing Frozen WavLM Feature Extractor...")
processor = Wav2Vec2FeatureExtractor.from_pretrained("microsoft/wavlm-base-plus")
wavlm_model = WavLMModel.from_pretrained("microsoft/wavlm-base-plus").to(device)
wavlm_model.eval()

print("🔥 Loading Version 8 XGBoost Meta-Classifier...")
xgb_model = xgb.XGBClassifier()
xgb_model.load_model(MODEL_PATH)
print("✅ Systems Online! Ready for testing.")

# ==========================================
# 2. IN-MEMORY PIPELINE FUNCTIONS
# ==========================================
def extract_master_dna(audio_path):
    """Processes full-length audio in memory with 16kHz sliding window"""
    waveform, sr = torchaudio.load(audio_path, backend="soundfile")
    
    # Absolute frequency normalization (forces model to look at biometrics, not sample rates)
    if sr != 16000:
        waveform = torchaudio.transforms.Resample(sr, 16000)(waveform)
        
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)
        
    waveform = waveform.squeeze().numpy()

    chunk_length = 16000 * 4  # 4 seconds
    stride = 16000 * 2        # 2 seconds overlap
    
    if len(waveform) <= chunk_length:
        chunks = [np.pad(waveform, (0, chunk_length - len(waveform)), "constant")]
    else:
        chunks = [waveform[i:i + chunk_length] for i in range(0, len(waveform) - chunk_length + 1, stride)]
        if len(waveform) % stride != 0 and len(waveform) > chunk_length:
            chunks.append(waveform[-chunk_length:])

    chunk_embeddings = []
    with torch.no_grad():
        for chunk in chunks:
            inputs = processor(chunk, sampling_rate=16000, return_tensors="pt").input_values.to(device)
            emb = wavlm_model(inputs).last_hidden_state.mean(dim=1).squeeze(0)
            chunk_embeddings.append(emb)

    master_dna = torch.stack(chunk_embeddings).mean(dim=0)
    return master_dna.cpu().numpy().reshape(1, -1)

def run_inference(change):
    output_area.clear_output()
    with output_area:
        if not uploader.value:
            print("⚠️ Please upload a valid .wav file first.")
            return
            
        # --- IPYWIDGETS 8.x FIX ---
        # uploader.value is now a tuple of dictionaries
        uploaded_file = uploader.value[0]
        filename = uploaded_file['name']
        content = uploaded_file['content']
        # --------------------------
        
        # Temporary drop file to disk for torchaudio to ingest safely
        temp_path = f"temp_eval_{filename}"
        with open(temp_path, "wb") as f:
            f.write(content)
            
        print(f"🔍 Analyzing clip: '{filename}' across full length...")
        
        try:
            # Run exact extraction pipeline
            dna_features = extract_master_dna(temp_path)
            
            # Predict probabilities
            fake_prob = xgb_model.predict_proba(dna_features)[0][1]
            real_prob = 1.0 - fake_prob
            
            # Format display metrics
            print("\n" + "="*50)
            print("         FORENSIC THREAT ANALYSIS REPORT        ")
            print("="*50)
            print(f"Authentic Human Speech Confidence : {real_prob * 100:.2f}%")
            print(f"Synthetic AI Vocoder Confidence   : {fake_prob * 100:.2f}%")
            print("="*50)
            
            if fake_prob > 0.85:
                display(HTML("<h2 style='color:crimson;'>🚨 CRITICAL THREAT: Confirmed Deepfake</h2>"))
            elif fake_prob > 0.50:
                display(HTML("<h2 style='color:orange;'>⚠️ HIGH RISK: Anomalous Synthetic Patterns Detected</h2>"))
            else:
                display(HTML("<h2 style='color:seagreen;'>✅ VERIFIED SAFE: Authentic Human Vocal Tract</h2>"))
                
        except Exception as e:
            print(f"❌ Error processing file: {str(e)}")
        finally:
            if os.path.exists(temp_path):
                os.remove(temp_path)

# ==========================================
# 3. INTERACTIVE JUPYTER WIDGETS
# ==========================================
uploader = widgets.FileUpload(accept='.wav', multiple=False, description="Upload Waveform")
output_area = widgets.Output()

# Link action logic
uploader.observe(run_inference, names='value')

display(HTML("<h3>🎤 Audio Forensics Gateway - Grand V8 Engine (WavLM + XGBoost Head)</h3>"))
display(uploader, output_area)
```

```python
import os
import torch
import torchaudio
import numpy as np
import lightgbm as lgb
from transformers import Wav2Vec2FeatureExtractor, WavLMModel
import ipywidgets as widgets
from IPython.display import display, HTML
import warnings

warnings.filterwarnings("ignore")

# ==========================================
# 1. SETUP ENVIRONMENT & LOAD MODELS
# ==========================================
MODEL_PATH = "v8_1_production/v8_1_lightgbm.txt"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"❌ Could not find LightGBM model at '{MODEL_PATH}'.")

print(f"⏳ Booting up Frozen WavLM Engine on {device}...")
processor = Wav2Vec2FeatureExtractor.from_pretrained("microsoft/wavlm-base-plus")
wavlm_model = WavLMModel.from_pretrained("microsoft/wavlm-base-plus").to(device)
wavlm_model.eval()

# Initialize LFCC Extractor for Micro-Spectral Artifacts
lfcc_transform = torchaudio.transforms.LFCC(
    sample_rate=16000,
    n_lfcc=40,
    speckwargs={"n_fft": 512, "hop_length": 256, "center": False}
).to(device)

print("🔥 Loading Version 8.1 LightGBM Meta-Classifier...")
lgb_model = lgb.Booster(model_file=MODEL_PATH)
print("✅ Systems Online! Ready for Dual-Branch testing.")

# ==========================================
# 2. IN-MEMORY PIPELINE (NO AUGMENTATION)
# ==========================================
def extract_v8_1_dna(audio_path):
    """Dual-Branch Extraction: WavLM (Rhythm) + LFCC (Math)"""
    waveform, sr = torchaudio.load(audio_path, backend="soundfile")
    
    if sr != 16000:
        waveform = torchaudio.transforms.Resample(sr, 16000)(waveform)
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)
        
    # --- BRANCH 1: LFCC MICRO-SPECTRAL MATH (80 Features) ---
    lfccs = lfcc_transform(waveform.to(device)).squeeze(0) # Shape: [40, Time]
    lfcc_mean = torch.mean(lfccs, dim=1) 
    lfcc_std = torch.std(lfccs, dim=1)   
    lfcc_vector = torch.cat((lfcc_mean, lfcc_std), dim=0) 
    
    # --- BRANCH 2: WAVLM PARALINGUISTICS (768 Features) ---
    waveform_np = waveform.squeeze().numpy()
    chunk_length = 16000 * 4  # 4 seconds
    stride = 16000 * 2        # 2 seconds overlap
    
    if len(waveform_np) <= chunk_length:
        chunks = [np.pad(waveform_np, (0, chunk_length - len(waveform_np)), "constant")]
    else:
        chunks = [waveform_np[i:i + chunk_length] for i in range(0, len(waveform_np) - chunk_length + 1, stride)]
        if len(waveform_np) % stride != 0 and len(waveform_np) > chunk_length:
            chunks.append(waveform_np[-chunk_length:])

    chunk_embeddings = []
    with torch.no_grad():
        for chunk in chunks:
            inputs = processor(chunk, sampling_rate=16000, return_tensors="pt").input_values.to(device)
            emb = wavlm_model(inputs).last_hidden_state.mean(dim=1).squeeze(0)
            chunk_embeddings.append(emb)

    wavlm_vector = torch.stack(chunk_embeddings).mean(dim=0)
    
    # --- FUSION: Combine Both Branches ---
    super_vector = torch.cat((wavlm_vector.cpu(), lfcc_vector.cpu()), dim=0)
    return super_vector.numpy().reshape(1, -1)

# ==========================================
# 3. INTERACTIVE UPLOADER & INFERENCE
# ==========================================
def run_inference(change):
    output_area.clear_output()
    with output_area:
        if not uploader.value:
            print("⚠️ Please upload a valid .wav file first.")
            return
            
        # IPyWidgets 8.x specific tuple extraction
        uploaded_file = uploader.value[0]
        filename = uploaded_file['name']
        content = uploaded_file['content']
        
        # Temporary drop file to disk for torchaudio to ingest safely
        temp_path = f"temp_eval_{filename}"
        with open(temp_path, "wb") as f:
            f.write(content)
            
        print(f"🔍 Analyzing clip: '{filename}' using Dual-Branch fusion...")
        
        try:
            # Extract 848-D Super-Vector
            dna_features = extract_v8_1_dna(temp_path)
            
            # Predict probabilities via LightGBM
            fake_prob = lgb_model.predict(dna_features)[0]
            real_prob = 1.0 - fake_prob
            
            # Format display metrics
            print("\n" + "="*50)
            print("      V8.1 FORENSIC THREAT ANALYSIS REPORT      ")
            print("="*50)
            print(f"Authentic Human Speech Confidence : {real_prob * 100:.2f}%")
            print(f"Synthetic AI Vocoder Confidence   : {fake_prob * 100:.2f}%")
            print("="*50)
            
            if fake_prob > 0.85:
                display(HTML("<h2 style='color:crimson;'>🚨 CRITICAL THREAT: Confirmed Deepfake (RVC / TTS Detected)</h2>"))
            elif fake_prob > 0.50:
                display(HTML("<h2 style='color:orange;'>⚠️ HIGH RISK: Anomalous Synthetic Phase Patterns Detected</h2>"))
            else:
                display(HTML("<h2 style='color:seagreen;'>✅ VERIFIED SAFE: Authentic Human Vocal Tract & Math</h2>"))
                
        except Exception as e:
            print(f"❌ Error processing file: {str(e)}")
        finally:
            if os.path.exists(temp_path):
                os.remove(temp_path)

# ==========================================
# 4. LAUNCH UI
# ==========================================
uploader = widgets.FileUpload(accept='.wav', multiple=False, description="Upload Waveform")
output_area = widgets.Output()

uploader.observe(run_inference, names='value')

display(HTML("<h3>🎤 Audio Forensics Gateway - Version 8.1 (WavLM + LFCC)</h3>"))
display(uploader, output_area)
```
